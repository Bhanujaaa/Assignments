```python

#1.a building for binary min heap
import heapq
import random
import time
import numpy as np


heap_arr=[]
avl_lowest_50=[]
def heapify(arr, N, i,heap_swap):
    smallest = i  # Initialize smallest as root

    l = 2 * i + 1  # left = 2*i + 1
    r = 2 * i + 2  # right = 2*i + 2

    if l>N or r>N:
        return heap_swap
    else:
        # If left child is smallest than root
        if l < N and arr[l] < arr[smallest]:
            smallest = l

        # If right child is larger than smallest so far
        if r < N and arr[r] < arr[smallest]:
            smallest = r

        # If smallest is not root
        if smallest != i:
            arr[i], arr[smallest] = arr[smallest], arr[i]
            heap_swap=heap_swap+1

            # Recursively heapify the affected sub-tree
            heap_swap=heapify(arr, N, smallest,heap_swap)
        return heap_swap


# Function to build a Max-Heap from the given array
def build_heap(arr, N,heap_swap):
    # Index of last non-leaf node
    startIdx = N // 2 - 1

    # Perform reverse level order traversal
    # from last non-leaf node and heapify
    # each node
    for i in range(startIdx, -1, -1):
        heap_swap=heapify(arr, N, i,heap_swap)
    return (heap_swap,arr)



#1. a building for AVL tree


class AVLNode:
    def __init__(self, value):
        self.value = value
        self.height = 1
        self.left = None
        self.right = None

def insert_avl(root, value):
    if root is None:
        return AVLNode(value), 0

    rotations = 0

    if value < root.value:
        root.left, rotations_left = insert_avl(root.left, value)
        rotations += rotations_left
    else:
        root.right, rotations_right = insert_avl(root.right, value)
        rotations += rotations_right

    root.height = 1 + max(get_height(root.left), get_height(root.right))

    balance = get_balance(root)

    if balance > 1 and value < root.left.value:
        return right_rotate(root), rotations + 1

    if balance < -1 and value > root.right.value:
        return left_rotate(root), rotations + 1

    if balance > 1 and value > root.left.value:
        root.left = left_rotate(root.left)
        return right_rotate(root), rotations + 2

    if balance < -1 and value < root.right.value:
        root.right = right_rotate(root.right)
        return left_rotate(root), rotations + 2

    return root, rotations

def get_height(node):
    return node.height if node else 0

def get_balance(node):
    return get_height(node.left) - get_height(node.right)

def left_rotate(node):
    y = node.right
    T = y.left

    y.left = node
    node.right = T

    node.height = 1 + max(get_height(node.left), get_height(node.right))
    y.height = 1 + max(get_height(y.left), get_height(y.right))

    return y

def right_rotate(node):
    x = node.left
    T = x.right

    x.right = node
    node.left = T

    node.height = 1 + max(get_height(node.left), get_height(node.right))
    x.height = 1 + max(get_height(x.left), get_height(x.right))

    return x

def build_avl(elements):
    root = None
    rotations = 0

    for element in elements:
        root, rotations_for_element = insert_avl(root, element)
        rotations += rotations_for_element

    return root, rotations

def generate_permutation(n):
    return random.sample(range(1, n+1), n)

def measure_time(build_function, elements):
    start_time = time.time()
    build_function(elements)
    end_time = time.time()
    return end_time - start_time



#lowest 50 AVL

avl_lowest_elements = []

def in_order_traversal(node):
    global avl_lowest_elements
    if node is not None:
        in_order_traversal(node.left)
        lowest_elements.append(node.value)
        in_order_traversal(node.right)

#random 50 AVL

random_elements = []
def report_random_numbers(node):
    global random_elements
    if node is not None:
        if len(random_elements) < 50:
            random_elements.append(node.value)
        elif random.randint(1, len(random_elements) + 1) <= 50:
            random_elements[random.randint(0, 49)] = node.value
        report_random_numbers(node.left)
        report_random_numbers(node.right)



#1.building splay tree
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.parent = None

class SplayTree:
    def __init__(self):
        self.root = None
        self.rotation_count = 0

    def left_rotate(self, x):
        self.rotation_count += 1
        y = x.right
        x.right = y.left
        if y.left:
            y.left.parent = x
        y.parent = x.parent
        if not x.parent:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y
        y.left = x
        x.parent = y

    def right_rotate(self, y):
        self.rotation_count += 1
        x = y.left
        y.left = x.right
        if x.right:
            x.right.parent = y
        x.parent = y.parent
        if not y.parent:
            self.root = x
        elif y == y.parent.left:
            y.parent.left = x
        else:
            y.parent.right = x
        x.right = y
        y.parent = x

    def splay(self, x):
        while x.parent:
            if not x.parent.parent:
                if x == x.parent.left:
                    self.right_rotate(x.parent)
                else:
                    self.left_rotate(x.parent)
            elif (x == x.parent.left) == (x.parent == x.parent.parent.left):
                if x.parent.left == x:
                    self.right_rotate(x.parent.parent)
                    self.right_rotate(x.parent)
                else:
                    self.left_rotate(x.parent.parent)
                    self.left_rotate(x.parent)
            else:
                if x.parent.left == x:
                    self.right_rotate(x.parent)
                    self.left_rotate(x.parent)
                else:
                    self.left_rotate(x.parent)
                    self.right_rotate(x.parent)

    def insert(self, data):
        new_node = Node(data)
        y = None
        x = self.root
        while x:
            y = x
            if data < x.data:
                x = x.left
            else:
                x = x.right
        new_node.parent = y
        if not y:
            self.root = new_node
        elif data < y.data:
            y.left = new_node
        else:
            y.right = new_node
        self.splay(new_node)

    def count_rotations(self):
        return self.rotation_count


    #splay 50 lowest
    def in_order_traversal(self, node, result):
      if node is not None:
          self.in_order_traversal(node.left, result)
          result.append(node.data)
          if len(result) >= 50:
              return
          self.in_order_traversal(node.right, result)

    def find_lowest_50(self):
        result = []
        self.in_order_traversal(self.root, result)
        return result[:50]


    #b ii splay-random 50
    def collect_random_sample(self, node, result, sample_size):
      if node is None:
          return

      # Recursively explore the tree in random order
      self.collect_random_sample(node.left, result, sample_size)
      if len(result) < sample_size:
          result.append(node.data)
      elif random.random() < sample_size / (len(result) + 1):
          # Replace an existing element with probability sample_size / (len(result) + 1)
          index_to_replace = random.randint(0, sample_size - 1)
          result[index_to_replace] = node.data
      self.collect_random_sample(node.right, result, sample_size)

    def find_random_50(self):
      result = []
      sample_size = 50
      self.collect_random_sample(self.root, result, sample_size)
      return result



def build_with_permutation(permutation):
        tree = SplayTree()
        for num in permutation:
            tree.insert(num)
        return tree




# Number of elements
n = 50000

# Number of iterations
num_iterations = 5

# Generate the initial list of elements
elements = list(range(1, n+1))
extra_numbers = random.sample(range(50001, 100001), 5000)

# Perform the operations and record times
heap_times_all = []
heap_swaps_all = []
avl_times_all = []

avl_rotations_list_all = []
splay_build_times_all = []
splay_tree_rotations_all = []

heap_time_lowest_50_all=[]
avl_time_lowest_50_all=[]
splay_time_lowest_50_all=[]

heap_time_random_50_all=[]
avl_time_random_50_all=[]
splay_time_random_50_all=[]

heap_time_insertion=[]
avl_time_insertion=[]
splay_time_insertion=[]

# Perform the operation 5 times
for set_num in range(1, num_iterations+1):
    print("ITERATION:",{set_num})
    random.shuffle(elements)

    #BUILDING OF DATA STRUCTURES

    heap_swap=0
    heap_data=elements.copy()
    # Build the binary min heap
    # Calculate the time taken
    start_time = time.time()
    heap_swap,heap_arr = build_heap(heap_data,len(heap_data),heap_swap)
    end_time = time.time()
    time_taken = end_time - start_time
    heap_times_all.append(time_taken)
    heap_swaps_all.append(heap_swap)


    # Print results for each set
    print(f"\nTime taken to build min heap is: {time_taken:.10f} seconds, swaps min heap: {heap_swap}")


    #for AVL
    start_time = time.time()
    avl_root, rotations = build_avl(elements)
    end_time = time.time()
    avl_rotations_list_all.append(rotations)
    # build_time = measure_time(build_avl, elements)
    avl_times_all.append(end_time - start_time)


    print(f"\nTime taken to build AVL: {end_time - start_time:.10f} seconds, Rotations in AVL: {rotations}")

    #for SPLAY TREE:
    start_time = time.time()
    splay_tree=build_with_permutation(elements)
    end_time = time.time()
    splay_build_times_all.append(end_time - start_time)
    splay_tree_rotations_all.append(splay_tree.count_rotations())


    print(f"\nTime taken to build splay tree: {end_time - start_time:.10f} seconds, Rotations in splay tree: {splay_tree.count_rotations()}")

    #LOWEST 50 NUMEBRS IN EACH DATA STRUCTURES

    #minheap-lowest 50
    start_time = time.time()
    lowest_50 = heapq.nsmallest(50, heap_arr)
    end_time = time.time()
    heap_time_lowest_50_all.append(end_time - start_time)
    print(f"\nLowest 50 elements in min heap: {lowest_50} , Time taken to find in min heap: {end_time - start_time:.10f} seconds")

    #AVL-lowest 50
    lowest_elements = []  # Reset the lowest elements list
    start_time = time.time()
    in_order_traversal(avl_root)
    end_time = time.time()
    avl_time_lowest_50_all.append(end_time - start_time)
    print(f"\nLowest 50 elements in AVL tree: {lowest_elements[:50]},Time taken to find in avl tree: {end_time - start_time:.10f} seconds")  # Print the lowest 50 elements


    #Splay-lowest 50
    start_time = time.time()
    lowest_50 = splay_tree.find_lowest_50()
    end_time = time.time()
    splay_time_lowest_50_all.append(end_time - start_time)
    print(f"\nLowest 50 elements in splay tree: {lowest_50[:50]}, Time taken to find in splay tree: {end_time - start_time:.10f} seconds")

    #min-heap random 50
    start_time = time.time()
    random_50 = random.sample(heap_arr, 50)
    end_time = time.time()
    heap_time_random_50_all.append(end_time - start_time)
    print(f"\nReported 50 random elements in min heap: {random_50},Time taken to report in min heap: {end_time - start_time:.10f} seconds")

    #avl-random 50
    random_elements = []  # Reset the random elements list
    start_time = time.time()
    report_random_numbers(avl_root)
    end_time = time.time()
    avl_time_random_50_all.append(end_time - start_time)
    print(f"\nReported 50 random elements in avl tree: {random_elements},Time taken to report in avl tree: {end_time - start_time:.10f} seconds")

    #splay-random 50
    random_elements = []
    start_time = time.time()
    random_elements = splay_tree.find_random_50()
    end_time = time.time()
    splay_time_random_50_all.append(end_time - start_time)
    print(f"\nReported 50 random elements in splay tree: {random_elements},Time taken to report in splay tree: {end_time - start_time:.10f} seconds")

    #1.c insertion - min heap
    start_time = time.time()
    # Insert the 5000 random numbers
    for num in extra_numbers:
        heapq.heappush(heap_arr, num)
    end_time = time.time()
    time_taken = end_time - start_time
    heap_time_insertion.append(time_taken)
    print(f"\nNumber of insertions: {len(extra_numbers)},Time taken to insert 5000 elements in min heap: {time_taken:.10f} second")


#1.c insertion - AVL
    start_time = time.time()
    # Insert the 5000 random numbers
    for num in extra_numbers:
        avl_root,rotations=insert_avl(avl_root,num)
    end_time = time.time()
    time_taken = end_time - start_time
    avl_time_insertion.append(time_taken)
    print(f"\nNumber of insertions: {len(extra_numbers)},Time taken to insert 5000 elements in AVL: {time_taken:.10f} second")


#1.c insertion - splay
    start_time = time.time()
    # Insert the 5000 random numbers

    for num in extra_numbers:
        splay_tree.insert(num)
    end_time = time.time()
    time_taken = end_time - start_time
    splay_time_insertion.append(time_taken)
    print(f"\nNumber of insertions: {len(extra_numbers)},Time taken to insert 5000 elements in splay tree: {time_taken:.10f} second")


    print("\n********************************","END OF ITERATION:",{set_num},"********************************")


# Perform the operation 5 times (for statistical calculations)
#MIN HEAP
# Calculate mean and standard deviation min heap
# Print mean and standard deviation for building min heap
heap_mean_time = np.mean(heap_times_all)
heap_std_dev_time = np.std(heap_times_all)
heap_mean_swaps = np.mean(heap_swaps_all)
heap_std_dev_swaps = np.std(heap_swaps_all)
print("\nMEAN AND STANDARD DEVIATION OF MIN HEAP")
print(f"Mean number of swaps in min heap: {heap_mean_swaps}, Standard deviation of swaps in min heap: {heap_std_dev_swaps}")
print(f"Mean time of min heap: {heap_mean_time} seconds, Standard deviation of time min heap: {heap_std_dev_time} seconds")

#print mean and std for finding lowest 50 and random 50 numebrs min heap
heap_mean_time_lowest_50 = np.mean(heap_time_lowest_50_all)
heap_std_dev_time_lowest_50 = np.std(heap_time_lowest_50_all)
heap_mean_time_random_50 = np.mean(heap_time_random_50_all)
heap_std_dev_time_random_50 = np.std(heap_time_random_50_all)
print(f"Mean time for finding lowest 50 min heap: {heap_mean_time_lowest_50} seconds, Standard deviation of time for lowest 50 min heap: {heap_std_dev_time_lowest_50} seconds")
print(f"Mean time for finding random 50 min heap: {heap_mean_time_random_50} seconds, Standard deviation of time for random 50 min heap: {heap_std_dev_time_random_50} seconds")


#print mean and std for insertion min heap
heap_mean_time_insertion = np.mean(heap_time_insertion)
heap_std_time_insertion = np.std(heap_time_insertion)
print(f"Mean time for inserting 5000 elements in min heap: {heap_mean_time_insertion} seconds, Standard deviation of time for inserting 5000 elements min heap: {heap_std_time_insertion} seconds")


#AVL
# Calculate mean and standard deviation for AVL
avl_mean_time = np.mean(avl_times_all)
avl_std_dev_time = np.std(avl_times_all)
avl_mean_rotations = np.mean(avl_rotations_list_all)
avl_std_dev_rotations = np.std(avl_rotations_list_all)

# Print mean and standard deviation of rotations and time
print("\nMEAN AND STANDARD DEVIATION OF AVL TREE")
print(f"Mean rotations of AVL: {avl_mean_rotations}, Standard deviation of rotations AVL: {avl_std_dev_rotations}")
print(f"Mean Time taken to build AVL: {avl_mean_time} seconds, Standard deviation Time taken to build AVL: {avl_std_dev_time} seconds")


#print mean and std for finding lowest 50 and random 50 AVL
avl_mean_time_lowest_50 = np.mean(avl_time_lowest_50_all)
avl_std_dev_time_lowest_50 = np.std(avl_time_lowest_50_all)
avl_mean_time_random_50 = np.mean(avl_time_random_50_all)
avl_std_dev_time_random_50 = np.std(avl_time_random_50_all)
print(f"Mean time for finding lowest 50 avl: {avl_mean_time_lowest_50} seconds, Standard deviation of time for lowest 50 avl: {avl_std_dev_time_lowest_50} seconds")
print(f"Mean time for finding random 50 avl: {avl_mean_time_random_50} seconds, Standard deviation of time for random 50 avl: {avl_std_dev_time_random_50} seconds")

#print mean and std for insertion AVL
avl_mean_time_insertion = np.mean(avl_time_insertion)
avl_std_time_insertion = np.std(avl_time_insertion)
print(f"Mean time for inserting 5000 elements in AVL: {avl_mean_time_insertion} seconds, Standard deviation of time for inserting 5000 elements AVL: {avl_std_time_insertion} seconds")

#SPLAY
# Calculate mean and standard deviation for build times and rotations for splay tree
splay_mean_time = np.mean(splay_build_times_all)
splay_std_dev_time = np.std(splay_build_times_all)
splay_mean_rotations = np.mean(splay_tree_rotations_all)
splay_std_dev_rotations = np.std(splay_tree_rotations_all)


print("\nMEAN AND STANDARD DEVIATION OF SPLAY TREE")
print(f"Mean Rotations in splay tree: {splay_mean_rotations}, Standard deviation Rotations in splay tree: {splay_std_dev_rotations}")
print(f"Mean Time taken to build splay tree: {splay_mean_time:.10f} seconds, Standard deviation Time taken to build splay tree: {splay_std_dev_time:.10f} seconds")

#print mean and std for finding lowest 50 and random 50 numbers splay
splay_mean_time_lowest_50 = np.mean(splay_time_lowest_50_all)
splay_std_dev_time_lowest_50 = np.std(splay_time_lowest_50_all)
splay_mean_time_random_50 = np.mean(splay_time_random_50_all)
splay_std_dev_time_random_50 = np.std(splay_time_random_50_all)
print(f"Mean time for finding lowest 50 splay tree: {splay_mean_time_lowest_50} seconds, Standard deviation of time for lowest 50 splay: {splay_std_dev_time_lowest_50} seconds")
print(f"Mean time for finding random 50 splay tree: {splay_mean_time_random_50} seconds, Standard deviation of time for random 50 splay: {splay_std_dev_time_random_50} seconds")


#print mean and std for insertion splay
splay_mean_time_insertion = np.mean(splay_time_insertion)
splay_std_time_insertion = np.std(splay_time_insertion)
print(f"Mean time for inserting 5000 elements in splay tree: {splay_mean_time_insertion} seconds, Standard deviation of time for inserting 5000 elements splay: {splay_std_time_insertion} seconds")
```
